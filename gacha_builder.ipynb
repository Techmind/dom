```python
import pandas as pd
```

```python
df = pd.read_csv('dom6inspector/gamedata/BaseU.csv', sep='\t')
df.head()
df = df.set_index('id')
df = df.fillna(0)

df['is_mage'] = df[['F','A','W','E','S','D','N','G','B','H']].sum(axis=1)
#df[['F','A','W','E','S','D','N','G','B','H']].loc[42]
df
```

```python
df_gold = pd.read_csv('gcost.csv', sep='\t')
df_gold.head()
df_gold.set_index('id')

df_full = pd.merge(df, df_gold, on='id', how='inner')
df_full = df_full.fillna(0)
df_full
```

```python
ids_mages = df_full.loc[(df_full['is_mage'] >= 1) & (df_full['is_mage'] <= 3) & (df_full['goldcost'] < 200) & (df_full['goldcost'] > 0)]['id'].to_list()

for i, num in enumerate(ids_mages):
    print(f"""#selectmonster {num}
#montag 1500
#addupkeep -1000
#end""")
    
ids_good_mages = df_full.loc[(df_full['is_mage'] >= 4) & (df_full['goldcost'] > 200)]['id'].to_list()

for i, num in enumerate(ids_good_mages):
    print(f"""#selectmonster {num}
#montag 1501
#addupkeep -1000
#end""")

    ids_commanders = df_full.loc[(df_full['leader'] >= 75) & (df_full['goldcost'] > 100)]['id'].to_list()

for i, num in enumerate(ids_good_mages):
    print(f"""#selectmonster {num}
#montag 1501
#addupkeep -1000
#end""")
    
    ids_commanders = df_full.loc[(df_full['leader'] >= 75) & (df_full['goldcost'] > 100) & (df_full['is_mage'] <= 2)]['id'].to_list()

for i, num in enumerate(ids_commanders):
    print(f"""#selectmonster {num}
#montag 1502
#addupkeep -1000
#end""")
```
