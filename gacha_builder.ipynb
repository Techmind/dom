```python
df['fake_prot'] = df.apply(lambda row: 10 if row['prot'] == 0 and (row['armor1'] != 0 or row['armor2'] != 0) else row['prot'], axis=1)
df['chasis_value'] = df['str']*df['hp']*((df['mr'])/10)*((df['fake_prot']+1)/10)*((df['att']+1)/12)*((df['def']+1)/12)
df['ehp'] = df['hp']*(df['fake_prot']/10)

df_sorted = df.sort_values('chasis_value', ascending=False)
#df_sorted.head(30)
```

```python
import pandas as pd

# Load the summoning spells data
summoning_spells_df = pd.read_csv('summoning_spells_output.csv', sep=',')

# Create a dictionary for quick lookup of unit summon costs
unit_gem_costs = {}
for _, row in summoning_spells_df.iterrows():
    unit_id = str(row['unit_id'])
    cnt = row['number_of_units']
    num_units = int(cnt % 100 + (cnt / 100 / 10 * 3 ))
    cost_in_gems = row['cost_in_gems']
    gempath = row['gempath']
    if num_units > 0:
        unit_gem_costs[unit_id] = {'cost' : cost_in_gems / num_units, 'path' : gempath}        
        
units_per_cast = {}
for _, row in summoning_spells_df.iterrows():
    unit_id = str(row['unit_id'])
    cnt = row['number_of_units']
    num_units = int(cnt % 100 + (cnt / 100 / 10 * 3 ))
    cost_in_gems = row['cost_in_gems']
    if num_units > 0:
        units_per_cast[unit_id] = num_units
                
# Function to get the gem cost for a unit
def get_unit_gem_cost(unit_id, unit_gem_costs):
    return unit_gem_costs.get(str(unit_id), {'cost' : 0})['cost']

def get_unit_gem_path(unit_id, unit_gem_costs):
    return unit_gem_costs.get(str(unit_id), {'path' : 0})['path']


def get_units_per_cast(unit_id, units_per_cast):
    cnt = units_per_cast.get(str(unit_id), 0)
    
    return cnt

print(unit_gem_costs)

df_sorted = df_sorted.loc[~df_sorted.index.duplicated(), :]

# Add a new column 'gem_cost' to the df_sorted dataframe
df_sorted['gem_cost'] = df_sorted.index.map(lambda unit_id: get_unit_gem_cost(unit_id, unit_gem_costs))
df_sorted['gem_path'] = df_sorted.index.map(lambda unit_id: get_unit_gem_path(unit_id, unit_gem_costs))
df_sorted['units_per_cast'] = df_sorted.index.map(lambda unit_id: get_units_per_cast(unit_id, units_per_cast))
```

```python
df_sorted['chassi_per_gem'] = df_sorted['chasis_value'] / df_sorted['gem_cost']
df_sorted['str_per_gem'] = (df_sorted['str'] ** 0.5 + df_sorted['siegebonus']) / df_sorted['gem_cost']
df_sorted['str_per_cast'] = (df_sorted['str'] ** 0.5 + df_sorted['siegebonus']) * df_sorted['units_per_cast']
df_sorted['hp_per_gem'] = (df_sorted['hp']) / df_sorted['gem_cost']
df_sorted['ehp_per_gem'] = (df_sorted['ehp']) / df_sorted['gem_cost']
df_sorted['ehp_per_cast'] = (df_sorted['ehp']) * df_sorted['units_per_cast']
df_sorted.head(20)
```

```python
#read magic_gem_income_per_location_per_level_MA_excluding_unique_False.json
import json
import pandas as pd

# Read the JSON file
with open('magic_gem_income_per_location_per_level_MA_excluding_unique_False.json', 'r') as f:
    map = json.load(f)

# Convert the nested dictionary (location -> school -> path -> value) to a DataFrame
records = []
for location, schools in map.items():
    for school, paths in schools.items():
        for path, value in paths.items():
            records.append({
                'location': location,
                'school': school,
                'path': path,
                'value': value
            })

gem_income_df = pd.DataFrame(records)

print(gem_income_df)

import itertools

# Get all unique paths in the data
unique_school = gem_income_df['school'].unique()

path_combs = itertools.combinations(unique_school, 2)

combination_results = []
combination_search_orders = []

# Iterate through all unique pairs of paths (order matters, skip same path)
for path1 in unique_school:
    for path2 in unique_school:
        if path1 == path2:
            continue
        for level1 in range(0, 5):
            for level2 in range(0, 5):
                # Build selectors for path-level pairs
                if 'level' in gem_income_df.columns:
                    sel1 = (gem_income_df['school'] == path1) & (gem_income_df['path'].astype(int) == level1)
                    sel2 = (gem_income_df['school'] == path2) & (gem_income_df['path'].astype(int) == level2)
                else:
                    sel1 = (gem_income_df['school'] == path1) & (gem_income_df['path'].astype(int) == level1)
                    sel2 = (gem_income_df['school'] == path2) & (gem_income_df['path'].astype(int) == level2)
                df1 = gem_income_df[sel1]
                df2 = gem_income_df[sel2]
                sum_val = df1['value'].sum() + df2['value'].sum()
                combination_label = f'{path1}{level1},{path2}{level2}'
                column1 = f'{path1}{level1}'
                column2 = f'{path2}{level2}'
                combination_results.append({
                    'combination': combination_label,
                    'column1': column1,
                    'column2': column2,
                    'paths': [path1, path2],
                    'levels': [level1, level2],
                    'sum_value': sum_val
                })

                # Location search order (from highest to lowest value sum for this combo)
                # Sum values per location, so merge per location over both df1 and df2
                df_combo = pd.concat([df1, df2])
                location_sums = df_combo.groupby('location')['value'].sum().sort_values(ascending=False)
                search_order = location_sums.index.tolist()
                combination_search_orders.append({
                    'combination': combination_label,
                    'path1': column1,
                    'path2': column2,
                    'search_order': search_order
                })

# Save overall pairwise path combinations to CSV
pairwise_df = pd.DataFrame(combination_results)
pairwise_df.to_csv('pairwise_gem_income_by_path_and_level.csv', index=False)

# Save location search orders for all combinations to CSV:
# Each row has: combination, search_order (as a comma-separated string)
search_order_df = pd.DataFrame([
    {
        "combination": row['combination'],
        'path1': row['path1'],
        'path2': row['path2'],
        "search_order": ",".join(row['search_order'])
    }
    for row in combination_search_orders
])
search_order_df.to_csv('path_search_order.csv', index=False)

```
