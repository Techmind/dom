```python
df = pd.read_csv('scbmBaseU.csv', sep=',')
df.head()
df = df.set_index('id')
df = df.fillna(0)

df['is_mage'] = df[['F','A','W','E','S','D','N','G','B','H']].sum(axis=1)
df[['F','A','W','E','S','D','N','G','B','H']].loc[42]
pd.set_option('display.max_columns', 20)
pd.set_option('display.max_rows', None)
df.loc[694]
```

```python
df['fake_prot'] = df.apply(lambda row: 10 if row['prot'] == 0 and (row['armor1'] != 0 or row['armor2'] != 0) else row['prot'], axis=1)
df['chasis_value'] = df['str']*df['hp']*(df['mr']/10)*(df['fake_prot']/10)*(df['att']/12)*(df['def']/12)
df['ehp'] = df['hp']*(df['fake_prot']/10)

df_sorted = df.sort_values('chasis_value', ascending=False)
#df_sorted.head(30)
```

```python
import pandas as pd

# Load the summoning spells data
summoning_spells_df = pd.read_csv('summoning_spells_scbm.csv', sep=',')

# Create a dictionary for quick lookup of unit summon costs
unit_gem_costs = {}
for _, row in summoning_spells_df.iterrows():
    unit_id = row['unit_id']
    num_units = row['number_of_units']
    cost_in_gems = row['cost_in_gems']
    if num_units > 0:
        unit_gem_costs[unit_id] = cost_in_gems / num_units
        
units_per_cast = {}
for _, row in summoning_spells_df.iterrows():
    unit_id = row['unit_id']
    num_units = row['number_of_units']
    cost_in_gems = row['cost_in_gems']
    if num_units > 0:
        units_per_cast[unit_id] = num_units
        

# Function to get the gem cost for a unit
def get_unit_gem_cost(unit_id, unit_gem_costs):
    return unit_gem_costs.get(str(unit_id), 0)

def get_units_per_cast(unit_id, units_per_cast):
    return units_per_cast.get(str(unit_id), 0)


df_sorted = df_sorted.loc[~df_sorted.index.duplicated(), :]

# Add a new column 'gem_cost' to the df_sorted dataframe
df_sorted['gem_cost'] = df_sorted.index.map(lambda unit_id: get_unit_gem_cost(unit_id, unit_gem_costs))
df_sorted['units_per_cast'] = df_sorted.index.map(lambda unit_id: get_units_per_cast(unit_id, units_per_cast))
```

```python
df_sorted['chassi_per_gem'] = df_sorted['chasis_value'] / df_sorted['gem_cost']
df_sorted['str_per_gem'] = (df_sorted['str'] ** 0.5 + df_sorted['siegebonus']) / df_sorted['gem_cost']
df_sorted['hp_per_gem'] = (df_sorted['hp']) / df_sorted['gem_cost']
df_sorted['ehp_per_gem'] = (df_sorted['ehp']) / df_sorted['gem_cost']
df_sorted['ehp_per_cast'] = (df_sorted['ehp']) * df_sorted['units_per_cast']
df_sorted.head(20)
```

```python
df_sorted = df_sorted.sort_values('ehp_per_cast', ascending=False)

df.loc[694]
# filter out inf
df_filtered = df_sorted.loc[~df_sorted.isin([np.inf, -np.inf]).any(axis=1)]

df_filtered[['name', 'gem_cost', 'fake_prot', 'mr', 'size', 'hp', 'ehp', 'chassi_per_gem', 'att', 'def', 'str', 'str_per_gem', 'hp_per_gem', 'ehp_per_gem', 'ehp_per_cast']].head(150)
```
